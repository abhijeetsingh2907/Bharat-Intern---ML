# WINE QUALITY PREDICTION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

```python
from sklearn.datasets import load_wine
wine_data=load_wine()
```

```python
input=wine_data.data
output=wine_data.target
column=wine_data.feature_names
target_names=wine_data.target_names
```

```python
df=pd.DataFrame(input, columns=column)
df.head()
```

```python
df1=pd.DataFrame(output)
df1["target"]=df1
df1=df1["target"]
df1.head()
```

```python
df=pd.concat([df,df1],axis=1)
df.head()
```

```python
df.info()
```

NONE OF THE FEATURE HAS OBJECT DATATYPE

```python
df.isnull().sum()
```

NO NULL VALUE

```python
df.describe()
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

```python
y=df["target"]
x=df.drop(["target"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=15)  
```

# CLASSIFICATION ALGORITHM

```python
from sklearn.tree import DecisionTreeClassifier
dtf=DecisionTreeClassifier()
```

```python
dtf=dtf.fit(x_train,y_train)
```

```python
y_pred=dtf.predict(x_test)
y_pred
```

```python
np.array(y_test)
```

# ACCURACY

```python
from sklearn.metrics import accuracy_score, classification_report, precision_recall_fscore_support
accuracy_score=accuracy_score(y_test,y_pred)
classification_report=classification_report(y_test,y_pred,target_names=target_names)
```
